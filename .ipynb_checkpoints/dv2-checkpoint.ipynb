<h3>1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset
np.random.seed(42)
x = np.random.rand(100) * 10  # Random values for X variable
y = 2.5 * x + np.random.randn(100) * 5  # Linear relationship with some noise

# Create a DataFrame
data = {
    'X': x,
    'Y': y
}
df = pd.DataFrame(data)

# Create a scatter plot using Seaborn
fig = sns.relplot(x=x, y=y , data=df)
```

<h3>2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a dataset of random numbers
np.random.seed(42)
data = np.random.randn(1000)  # 1000 random numbers from a normal distribution

# Visualize the distribution of the numerical variable using Seaborn
sns.histplot(data, kde = True)
plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

<h3>3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataset representing categories and their corresponding values
np.random.seed(42)
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))

# Create a DataFrame
data = {
    'Category': categories,
    'Value': values
}
df = pd.DataFrame(data)

# Compare different categories based on numerical values using a bar plot
sns.barplot(x='Category', y='Value', data=df)
plt.title('Comparison of Categories Based on Values')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

<h3>4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a dataset with categories and numerical values
np.random.seed(42)
categories = np.random.choice(['Category A', 'Category B', 'Category C', 'Category D'], 200)
values = np.random.randn(200) * 10 + 50  # Normal distribution with mean=50 and std=10

# Create a DataFrame
data = {
    'Category': categories,
    'Value': values
}
df = pd.DataFrame(data)

# Visualize the distribution of a numerical variable across different categories using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Distribution of Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

<h3>5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with simpler correlated features
n_samples = 100
feature_1 = np.random.randn(n_samples)
feature_2 = 0.8 * feature_1 + np.random.randn(n_samples) * 0.2
feature_3 = 0.5 * feature_1 + np.random.randn(n_samples) * 0.5

# Create a DataFrame
data = {
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3
}
df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr() 

# Visualize the correlation matrix using a heatmap
sns.heatmap(corr_matrix, linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

<h3>3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.


```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

output_file('fruits.html')
p = figure(x_range=fruits, title='BAR CHART')  # Set x_range to the categorical values
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Counts"
p.vbar(x=fruits, top=counts, width=0.5)  # Correct usage of x and top

show(p)
```

<h3>4. Create a Bokeh histogram to visualize the distribution of the given data.


```python
import numpy as np
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show

# Given data
data_hist = np.random.randn(1000)

# Generate histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Create the Bokeh plot
output_file('histogram.html')
p = figure(title='Histogram of Random Data', background_fill_color="#fafafa")

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white")

p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'

# Show the plot
show(p)
```

<h3>5. Create a Bokeh heatmap using the provided dataset.


```python
import numpy as np
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
from bokeh.colors import RGB

# Provided dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create the Bokeh plot
output_file('heatmap.html')
p = figure(title='Heatmap', x_range=(0,1), y_range=(0,1), tools='', toolbar_location=None)

# Since the data is small (10x10), we'll use `image` with dimensions scaled to match x and y
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, palette="Viridis256")

# Add grid lines and labels
p.grid.grid_line_width = 0.5
p.xaxis.axis_label = 'X-axis'
p.yaxis.axis_label = 'Y-axis'

# Show the plot
show(p)
```
