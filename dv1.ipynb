<h3>2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Generate the student grades data
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

# Create the violin plot using Plotly
fig = px.violin(df, x="Grade", y="Score",
                title="Distribution of Scores Across Grade Categories",
                labels={"Grade": "Grade Category", "Score": "Student Scores"},
                color="Grade")

# Show the violin plot
fig.show()
```

<h3>3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Generate the sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix format for the heatmap
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum', fill_value=0)

# Create the heatmap using Plotly
fig = px.imshow(pivot_table, 
                labels=dict(x="Month", y="Day", color="Sales"),
                x=pivot_table.columns,
                y=pivot_table.index,
                color_continuous_scale='YlGnBu',
                aspect="auto")

fig.update_layout(title='Sales Variation Across Months and Days')

# Show the heatmap
fig.show()
```

<h3>4. Using the given x and y data, generate a 3D surface plot to visualize the function 

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data and create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z)])

# Customize the layout
fig.update_layout(title='3D Surface Plot of the Function',
                  scene=dict(
                      xaxis_title='X Axis',
                      yaxis_title='Y Axis',
                      zaxis_title='Z Axis'),
                  width=800, height=600)

# Show the 3D surface plot
fig.show()
```

<h3>5. given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.

```python
import numpy as np
import pandas as pd
import plotly.express as px

# Seed for reproducibility
np.random.seed(25)

# Generating data
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Creating a bubble chart using Plotly
fig = px.scatter(df, x='GDP', y='Population', size='Population', hover_name='Country', 
                 title='Bubble Chart: Population vs. GDP',
                 labels={'GDP': 'GDP (in billion USD)', 'Population': 'Population (in millions)'},
                 size_max=60)

# Show the plot
fig.show()
```
